```python
import shutil
from pathlib import Path

# ===== CONFIG =====
SOURCE_DIR = Path("/Users/apple/Documents/obsidian/system design")       
TARGET_DIR = Path("")      
DOCS_DIR = TARGET_DIR / "docs"
SITE_NAME = "SVG Notes"
# ==================

def create_md_file(svg_path: Path, md_path: Path):
    title = svg_path.stem.replace("-", " ").replace("_", " ").title()
    content = f"""# {title}

![]({svg_path.name})
"""
    md_path.write_text(content, encoding="utf-8")


def copy_svgs_and_create_md():
    for svg in SOURCE_DIR.rglob("*.svg"):
        relative_path = svg.relative_to(SOURCE_DIR)
        target_svg_path = DOCS_DIR / relative_path
        target_md_path = target_svg_path.with_suffix(".md")

        # Create directories
        target_svg_path.parent.mkdir(parents=True, exist_ok=True)

        # Copy SVG
        shutil.copy2(svg, target_svg_path)

        # Create MD file
        create_md_file(target_svg_path, target_md_path)



def main():
    DOCS_DIR.mkdir(parents=True, exist_ok=True)
    copy_svgs_and_create_md()
    print("✅ SVGs copied, markdown files created, mkdocs.yml generated")


if __name__ == "__main__":
    main()
```
